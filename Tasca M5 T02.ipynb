### Train - Test

El train-test split divideix el dataset en dos per evaluar el rendiment d'un model de machine learning. Si utilitzes tot el teu conjunt de dades per a entrenar el model, és possible que s'ajusti en excés a les dades d'entrenament, fet que podria funcionar molt bé amb les dades que ja tens, però no amb les noves. 

La divisió en conjunts d'entrenament (train) i prova (test) permet simular com es comportarà el model en situacions reals. Típicament, s'utilitza una major proporció de dades pel train (ex: 80%) i una menor pel test (20%).

Donat el dataset, el que s'intentarà predir és la variable MEDV. Aquesta variable representa el valor dels habitatges, que és una de les mètriques més rellevants en un problema de preus d'habitatges (Variable objectiu (y)).

Les altres variables es poden utilitzar com a predictors per a entrenar el model (Variables de característiques (X)). Aquestes variables representen diferents atributs dels habitatges o de l'entorn.

```python
from sklearn.model_selection import train_test_split

# Farem un train-test split 80-20 (amb random_state per obtenir sempre el mateix resultat)

X = df.drop('MEDV', axis=1)
y = df['MEDV']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

print(f'La mostra de X per TRAIN (80%) és de {len(X_train)} registres')
print(f'La mostra de X per TEST (20%) és de {len(X_test)} registres')
print (' ')
print(f'La mostra de y per TRAIN (80%) és de {len(y_train)} registres')
print(f'La mostra de y per TEST (20%) és de {len(y_test)} registres')
```

Ja hem vist el tamany de cada mostra, ara analitzarem les dades que tenim (juntem la X i la y per facilitar l'estudi i comparació dels datasets).

```python
train = pd.concat([X_train, y_train], axis=1)
test = pd.concat([X_test, y_test], axis=1)
```

```python
train.describe().round(2)
```

```python
test.describe().round(2)
```

Les mitjanes i les desviacions estàndard dels dos datasets són similars, i per tant sembla que la distribució entre train i test és correcta. Tot i això, farem un Test d'Hipòtesis per comprovar matemàticament que no existeixen diferències significatives entre el train i el test. 

Primer mirem si les variables dels dos datasets (train i test) tenen una distribució normal.

Hipòtesis:

- H0 (nul·la): La mostra té una distribució normal.

- H1 (alternativa): La mostra no té una distribució normal.

- Alpha (nivell de significança) = 0.05 (5%).

```python
# Test de normalitat de Shapiro-Wilk pel dataset TRAIN

from scipy.stats import shapiro

for variable in train:
    stat, p = shapiro(train[variable])
    
    print(variable)
    print('stat=%.3f, p=%.3f' % (stat, p))
    
    if p > 0.05:
        print('La distribució probablement és gaussiana')
    else:
        print('La distribució probablement NO és gaussiana')
    print(' ')
```

Cap de les variables té una distribució normal.

```python
# Test de normalitat de Shapiro-Wilk pel dataset TEST

from scipy.stats import shapiro

for variable in test:
    stat, p = shapiro(test[variable])
    
    print(variable)
    print('stat=%.3f, p=%.3f' % (stat, p))
    
    if p > 0.05:
        print('La distribució probablement és gaussiana')
    else:
        print('La distribució probablement NO és gaussiana')
    print(' ')
```

Cap de les variables té una distribució normal.

Seguidament, mirarem matemàticament si hi ha diferències significatives entre TRAIN i TEST. Com cap de les variables té una distribució normal, realitzarem un Test de Wilcoxon-Mann-Whitney. Fem servir aquesta prova perquè no tenim el mateix nombre de registres entre train (404) i test (101), i llavors el test de Spearman ens dona error.

Hipòtesis:

- H0: No hi ha diferències significatives.

- H1: Hi ha diferències significatives.

- Alpha: 0.05

```python
# Test de Wilcoxon-Mann-Whitney per TRAIN-TEST

from scipy.stats import mannwhitneyu

for variable in train.columns:
    statistic, p_value = mannwhitneyu(train[variable], test[variable])
    
    print(variable)
    print('statistic=%.1f, p-value=%.3f' % (statistic, p_value))
    
    if p_value < 0.05:
        print('Hi ha diferències significatives')
    else:
        print('No Hi ha diferències significatives')
        print(' ')
```

Efectivament, tal com havíem pensat, no hi ha diferències significatives entre train i test. Per tant, l'split s'ha realitzat correctament.
